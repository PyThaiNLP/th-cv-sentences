# Generating fake data

สร้างข้อมูลสมมติ

## ชื่อคน

```python
from faker import Faker

fake = Faker("th_TH")

for _ in range(50):
    print(fake.name())
```

## จำนวน

```python
from random import randrange, random

from pythainlp.util import bahttext, num_to_thaiword

for i in range(50):
    print(num_to_thaiword(randrange(1, 100)))
for _ in range(15):
    print(num_to_thaiword(randrange(101, 1000)))
for _ in range(10):
    print(num_to_thaiword(randrange(1001, 10000)))
for _ in range(10):
    print(num_to_thaiword(randrange(10001, 100000)))
for _ in range(5):
    print(num_to_thaiword(randrange(100001, 1000000)))
for _ in range(5):
    print(num_to_thaiword(randrange(1000001, 10000000)))
for _ in range(5):
    print(num_to_thaiword(randrange(10000001, 100000000)))
for _ in range(5):
    print(num_to_thaiword(randrange(100000001, 1000000000)))

for _ in range(20):
    print(bahttext(randrange(10000)))
for _ in range(20):
    print(bahttext(random() * (10 ** randrange(6))))
```
